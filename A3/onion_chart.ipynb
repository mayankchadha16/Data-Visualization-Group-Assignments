```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Electric_Vehicle_Title_and_Registration_Activity.csv")
```

```python
make_counts = df.groupby('Make')['DOL Vehicle ID'].nunique()
```

```python
make_counts
```

```python
top_5_makes = make_counts.sort_values(ascending=False).head(5)
```

```python
top_5_df = pd.DataFrame({'Make': top_5_makes.index, 'Count of cars': top_5_makes.values})
```

```python
top_5_df.head()
```

```python
top_5_df.to_csv('top_5_makes.csv', index=False)
```

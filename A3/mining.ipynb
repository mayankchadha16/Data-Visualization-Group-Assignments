```python
import seaborn as sns
import matplotlib.pyplot as plt
import statistics

# Assuming 'Make' is your object data type column
# Replace 'Make' with the actual name of your column

# Identify unique values in 'Make' column
unique_values = df2['Clean Alternative Fuel Vehicle Type'].unique()

# Split unique values into chunks of 5
chunks_of_5 = [unique_values[i:i+2] for i in range(0, len(unique_values), 2)]

# Create separate violin plots for each chunk
for chunk in chunks_of_5:
    plt.figure(figsize=(12, 6))
    subset_df = df2[df2['Clean Alternative Fuel Vehicle Type'].isin(chunk)]
    sns.violinplot(x='Clean Alternative Fuel Vehicle Type', y='Sale Price', data=subset_df, palette='muted')  
    # Replace 'NumericColumn' with the corresponding numerical column for the y-axis

    # Calculate and print statistics
    median_value = statistics.median(subset_df['Sale Price'])
    mode_value = statistics.mode(subset_df['Sale Price'])
    mean_value = statistics.mean(subset_df['Sale Price'])

    print(f"Subset: {', '.join(chunk)}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
    print(f"Mean: {mean_value}")
    print("="*30)

    # Add labels and title
    plt.xlabel('Make')
    plt.ylabel('Numeric Column')
    plt.title('Violin Plot of Make (Subset)')

    # Show the plot
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics

# Assuming 'Make' is your object data type column
# Replace 'Make' with the actual name of your column
# Replace 'NumericColumn' with the corresponding numerical column

# Create a DataFrame with the mean, mode, and median values for each 'Make'
make_stats = []

for make_value in df2['Clean Alternative Fuel Vehicle Type'].unique():
    subset_df = df2[df2['Clean Alternative Fuel Vehicle Type'] == make_value]
    
    # Calculate statistics
    median_value = statistics.median(subset_df['Sale Price'])
    mode_value = statistics.mode(subset_df['Sale Price'])
    mean_value = statistics.mean(subset_df['Sale Price'])
    
    make_stats.append({
        'Clean Alternative Fuel Vehicle Type': make_value,
        'Median': median_value,
        'Mode': mode_value,
        'Mean': mean_value
    })

# Convert the results to a DataFrame
stats_df = pd.DataFrame(make_stats)

# Print the statistics
print(stats_df)

# Plot violin plot for the entire 'Make' column
plt.figure(figsize=(12, 6))
sns.violinplot(x='Clean Alternative Fuel Vehicle Type', y='Sale Price', data=df2, palette='muted')

# Add labels and title
plt.xlabel('Clean Alternative Fuel Vehicle Type')
plt.ylabel('Sale Price')
plt.title('Violin Plot of Make')

# Show the plot
plt.show()
```

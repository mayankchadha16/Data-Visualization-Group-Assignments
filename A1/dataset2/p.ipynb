```python
df[(df["New or Used Vehicle"] == "New") & (df["Make"] == "FORD")]["Model"].value_counts()
```

```python
temp1 = df[((df["Electric Range"] > 0) & (df["New or Used Vehicle"] == "New"))].groupby("Make")["Electric Range"].mean()
temp1
temp1.to_csv("electric_range_mean.csv");
```

```python
temp2 = df[((df["Base MSRP"] > 0) & (df["New or Used Vehicle"] == "New"))].groupby("Make")["Base MSRP"]
temp2
```

```python
temp2 = df[((df["Sale Price"] > 0) & (df["New or Used Vehicle"] == "New"))].groupby("Make")["Sale Price"].apply(lambda x: x.value_counts().idxmax())
temp2.to_csv("sale_price_mean.csv");
```
